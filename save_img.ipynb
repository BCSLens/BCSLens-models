```python
import os
import json
import requests

API_KEY = "AIzaSyCkiR8CUDdYAhVAtBOnRtEk8ShJlHqo4aM"   # üëà ‡πÉ‡∏™‡πà API Key ‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏∏‡∏ì‡∏ï‡∏£‡∏á‡∏ô‡∏µ‡πâ

json_file = "bcs_dataset.json"
output_json_file = "bcs_dataset_with_filenames.json"

# ‡πÇ‡∏´‡∏•‡∏î JSON
with open(json_file, "r", encoding="utf-8") as f:
    data = json.load(f)

image_fields = ["‡∏†‡∏≤‡∏û‡∏î‡πâ‡∏≤‡∏ô‡∏ö‡∏ô (Top View)", "‡∏†‡∏≤‡∏û‡∏î‡πâ‡∏≤‡∏ô‡∏´‡∏•‡∏±‡∏á (Back View)", 
                "‡∏†‡∏≤‡∏û‡∏î‡πâ‡∏≤‡∏ô‡∏Ç‡∏ß‡∏≤ (Right View)", "‡∏†‡∏≤‡∏û‡∏î‡πâ‡∏≤‡∏ô‡∏ã‡πâ‡∏≤‡∏¢ (Left View)"]

def get_file_name(file_id):
    url = f"https://www.googleapis.com/drive/v3/files/{file_id}?fields=name&key={API_KEY}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json().get("name")
    else:
        print(f"‚ùå Failed to get name for {file_id}: {response.text}")
        return None

for entry in data:
    for field in image_fields:
        if field in entry and entry[field].startswith("https://drive.google.com"):
            url = entry[field]
            file_id = url.split("id=")[-1]
            file_name = get_file_name(file_id)
            if file_name:
                entry[field] = file_name
            else:
                entry[field] = None

# ‡πÄ‡∏ã‡∏ü JSON ‡πÉ‡∏´‡∏°‡πà
with open(output_json_file, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print(f"‚úÖ Saved JSON with Google Drive file names: {output_json_file}")
```
